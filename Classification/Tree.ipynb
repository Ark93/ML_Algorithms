This is a Jupyter Ipython Notebook for Tree Decision Algorithm on Scykit Learn. Tree Decision Algorithm choose one features (data) per branch to make binary desicion in order to get the largest "information gain" buck using binary variables or numerical values more than and less than. Usually, for categorical variables is recomended to do a preprocessing in order to code the categorical variables into binary boolean variables. One example of this kind of algorithms is the OneHotEncoder. In this example I only use two features in order to make a surface plot, but it can be expanded to many features.

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
import numpy as np
from sklearn import datasets, preprocessing, cross_validation, metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
def plot_decision_regions(X, y, classifier, test_idx=None, resolution = 0.02):
    # marker generator and color map
    markers = ('s','x','o','^','v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() -1, X[:,0].max()+1
    x2_min, x2_max = X[:, 1].min() -1, X[:,0].max()+1
    xx1, xx2 = np.meshgrid(np.arange(x1_min,x1_max,resolution),np.arange(x2_min,x2_max,resolution))
    Z = classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1,xx2,Z,alpha=0.4,cmap = cmap)
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())
    
    # plot the samples
    X_test, y_test = X[test_idx,:], y[test_idx]
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl,0], y=X[y==cl,1], alpha = 0.8, c=cmap(idx), marker = markers[idx], label=cl)
    
    # highlight test samples
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:,1], c = '', alpha = 1.0, linewidth=1, marker='o', s=55, label='test set')
        
        
```

```python
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.3, random_state = 0)
```

```python
tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
tree.fit(X_train, y_train)
```

```python
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))
```

```python
plot_decision_regions(X_combined, y_combined, classifier=tree, test_idx=range(105,150))
plt.ylim((0,3))
plt.xlabel('petal Length [cm]')
plt.ylabel('petal Width [cm]')
plt.legend(loc='upper left')
plt.show()
```
